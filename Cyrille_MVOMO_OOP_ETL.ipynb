```python
# CEll class #
class CYPartnersETL():
  """
    This is a class for my OOP Project.
    
    Attributes:
      dict_data: Dictionnary of many differents data related to CYPARTNERS clients that I will usse for my OOP project

  """
  # Constructor #
  def __init__(self):
    """
      A constructor of a class is a special method that gets called when a class is instantiated using the NEW function. 
      A constructor for a class has the same name as the class name. Unlike ordinary methods, 
      a constructor definition is identified by the CONSTRUCTOR statement. In addition, constructors cannot have a return type.

      Parameters: 
        dict_data: Dictionnary of many differents data that I will use for my OOP project

    """
    self.dict_data = {
        'Gender':None,
        'age': 0,
        'name': None,
        'birthdate': None,
        'country': None,
        'continent':None,
        'company': None,
        'sport': None,
        'total earnings (M€)': 0,
        'on-field earnings (M€)': 0,
        'average off-field earnings before Cy (M€)': 0,
        'off-field earnings with Cy (M€)': 0,
        'off-field earnings increase rate (%)': 0,
        'Number of Lawsuits': 0,
        'Number of Instagram followers (M)': 0,
        'Average Number of Instagram followers increase rate (%)': 0,
        'Number of Instagram followers increase rate Before CY (%)': 0,
        'Number of Instagram followers increase rate With CY (%)': 0,
        'Number of meetings' : 0,
        'Legal expenses before CY (M€)': 0,
        'Legal expenses after CY (M€)': 0,
        'Money Generated (M€)': 0,
        'initial investment (M€)': 0,
        'Profitability': None,
        'ROI (%)': 0
        }


# One Static Methods #
  @staticmethod
  def get_inclusive_gender(gender_of_the_client):
    """
      The Function to turn all the genders into a neutral gender called 'M' in order to make our report more 'inclusive'
      
      Parameters:
        gender_of_the_client: The gender of the client that we want to change (can be 'Mr' or 'Miss')
          
      Returns:
        'M': A string which contains a neutral letter (M) depending on the conditions.
    """
    if gender_of_the_client == 'Mr':
      return 'M'
    return 'M'



# Mandatory Method n1 #
  def data_extracting(self, form):
    """
      The Function extracting informations from the form and making it attributes

      Parameters:
        form: The txt. file (called cyrille_form.txt) that my boss gave me for the task which contains clients informations
          
      Returns:
        A specific information from the form (e.g. the name of the client, or his Company).
    """
    self.dict_data['Gender'] = self.get_inclusive_gender(form.split()[0])
    self.dict_data['name'] = form.split()[1].split('.')[1]
    self.dict_data['birthdate'] = form.split()[4]
    self.dict_data['country'] = form.split()[7]
    self.dict_data['company'] = form.split()[10]
    self.dict_data['sport'] = form.split()[13]
    self.dict_data['total earnings (M€)'] = float(form.split()[19])
    self.dict_data['on-field earnings (M€)'] = float(form.split()[26])
    self.dict_data['Number of Lawsuits'] = int(form.split()[31])
    self.dict_data['Number of Instagram followers (M)'] = float(' '.join(
        form.split()[38:]
    ).replace('\n', ''))



  # Mandatory Method n2 #
  def data_featuring(self):
    """
      The Function featuring my attributes to create new data

      Parameters:
        any data extracted from my the form
          
      Returns:
        A new data (e.g. from birthdate, we got the age or from the country, we got the country code and then the continent)
    """    
    self.dict_data['birthdate'] = datetime.date(int(self.dict_data['birthdate'].split('/')[0]), 
    int(self.dict_data['birthdate'].split('/')[1]), 
    int(self.dict_data['birthdate'].split('/')[2])
    )
    self.dict_data['age'] = datetime.date.today().year - self.dict_data['birthdate'].year
    self.dict_data['off-field earnings (M€)'] = (self.dict_data['total earnings (M€)']) - (self.dict_data['on-field earnings (M€)']) #(i.e. at the moment when he joined CY)
    self.dict_data['country code'] = pc.country_name_to_country_alpha2(self.dict_data['country'], cn_name_format="default")
    self.dict_data['continent'] = pc.country_alpha2_to_continent_code(self.dict_data['country code'])
    
 
    

# Economical part: 3 customs methods #
  def average_off_field_earnings_BeforeCY(self):
    """
      The Function computing the value of the average off-field earning of each clients, before joining CY

      Parameters:
        the value of the off-field earnings when the client joined CY (obtained with data_featuring), 
        and a random integer between 1 and 3
          
      Returns:
        the result of the division between the value of the off-field earnings when the client joined CY,
        and a random integer between 1 and 3
    """   
    self.dict_data['average off-field earnings before Cy (M€)'] = (self.dict_data['off-field earnings (M€)']) / (random.randint(1, 3))#(i.e. before joining CY)
  def off_field_earnings_WithCY(self):
    """
      The Function computing the value of the average off-field earning of each clients, after joining CY

      Parameters:
        the value of the off-field earnings when the client joined CY (obtained with data_featuring), 
        and a random integer between 1 and 3
          
      Returns:
        the result of the multiplication between the value of the off-field earnings when the client joined CY,
        and a random integer between 1 and 3, minus 10% of the value of the off-field 
        earnings when the client joined CY (as concluded between CY and the client)
    """ 
    self.dict_data['off-field earnings with Cy (M€)'] = ((self.dict_data['off-field earnings (M€)']) * (random.randint(1, 3))) - (self.dict_data['off-field earnings (M€)'] * 0.1) #(i.e. after joining CY)
  def off_field_earnings_increase_rate(self):
    """
      The Function computing the value of the increase rate off-field earning of each clients, after joining CY

      Parameters:
        the value of the off-field earnings with CY and before CY
          
      Returns:
        the increase rate with the formula : ((final - initial) / initial) * 100
    """ 
    self.dict_data['off-field earnings increase rate (%)'] = (((self.dict_data['off-field earnings with Cy (M€)']) - 
    (self.dict_data['average off-field earnings before Cy (M€)'])) / (self.dict_data['average off-field earnings before Cy (M€)'])) *100 


  # Media part: 3 customs methods #
  def set_Average_numb_of_insta_followers_increase_rate(self, new_avg_numb):
    """
      The Function freezing the value of the Average Number of Instagram followers increase rate of each clients
      
      Parameters:
      The new average number

      Returns:
        the Average Number of Instagram followers increase rate
    """ 
    self.dict_data['Average Number of Instagram followers increase rate (%)'] = new_avg_numb
  def Numb_of_insta_follower_increase_rate_Before_CY(self):
    """
      The Function computing the value of the Number of Instagram followers increase rate of each clients, before joining CY

      Parameters:
        Average Number of Instagram followers increase rate and a random integer between 1 and 2
          
      Returns:
        The result of the multiplication between the Average Number of Instagram followers increase rate and a random integer between 1 and 2
    """ 
    self.dict_data['Number of Instagram followers increase rate Before CY (%)'] = (random.randint(1, 2)) * self.dict_data['Average Number of Instagram followers increase rate (%)']
  def Numb_of_insta_follower_increase_rate_with_CY(self):
    """
      The Function computing the value of the Number of Instagram followers increase rate of each clients, after joining CY

      Parameters:
        Number of Instagram followers increase rate Before CY and a random integer between 2 and 5
          
      Returns:
        The result of the multiplication between the Number of Instagram followers increase rate Before CY and a random integer between 2 and 5
    """ 
    self.dict_data['Number of Instagram followers increase rate With CY (%)'] = (random.randint(2, 5)) * self.dict_data['Number of Instagram followers increase rate Before CY (%)']

# Let's Set the Number Of Meetings: 1 customs methods    
  def set_Number_of_Meetings(self, new_number_of_meetings):
    """
      The Function freezing the value of the Number of meetings of each clients
      
      Parameters:
      The new Number of meetings

      Returns:
        the Number of meetings
    """ 
    self.dict_data['Number of meetings'] = new_number_of_meetings


# Legal part: 2 customs methods #
  def legal_expenses_before_CY(self):
    """
      The Function computing the Legal expenses before CY of each clients

      Parameters:
        Number of Lawsuits and 0.01 millions of euros (i.e. 10 000 euros (cost of lawsuits))
          
      Returns:
        The result of the multiplication between the Number of Lawsuits and 0.01
    """ 
    self.dict_data['Legal expenses before CY (M€)'] = 0.01 * self.dict_data['Number of Lawsuits']
  def legal_expenses_after_CY(self):
    """
      The Function computing the Legal expenses after joining CY of each clients

      Parameters:
        Legal expenses before CY
          
      Returns:
        The result of the division between the Legal expenses before CY and 2
    """ 
    self.dict_data['Legal expenses after CY (M€)'] = (self.dict_data['Legal expenses before CY (M€)']) / 2


# Let's add some interesting informations for the potential investors: 4 customs methods #
  def set_The_initial_Cost(self, new_cost):
    """
      The Function freezing the value of the initial investment (made by CYPARTNERS) for each clients
      
      Parameters:
      the new cost (the value of the investment)

      Returns:
        the initial investment
    """ 
    self.dict_data['initial investment (M€)'] = new_cost
  def has_generated(self):
    """
      The Function computing the Money Generated (by CYPARTNERS) by clients

      Parameters:
        off-field earnings with Cy, initial investment, Number of meetings and 0.3 (i.e. 300 000 euros which
        represent the fees of the collaborators)
          
      Returns:
        the result of the subtraction between off-field earnings with Cy, initial investment and 0.3 times the number
        of meetings
    """
    self.dict_data['Money Generated (M€)'] = (self.dict_data['off-field earnings with Cy (M€)'] * 0.1) - self.dict_data['initial investment (M€)'] - (self.dict_data['Number of meetings'] * 0.3) 
  def profitability(self):
    """
      The Function stating about the profitability of CYPARTNERS for each clients

      Parameters:
        Money Generated, initial investment
          
      Returns:
        String stating if it's a bad, good or excellent deal depending on the conditions 
    """
    #When
    if self.dict_data['Money Generated (M€)'] > self.dict_data['initial investment (M€)']:
      self.dict_data['Profitability'] = 'Good Deal'
    #When
    if self.dict_data['Money Generated (M€)'] > 2 * self.dict_data['initial investment (M€)']:
      self.dict_data['Profitability'] = 'Excellent Deal'
      #Then
      return self.dict_data['Profitability']
    #When
    if self.dict_data['Money Generated (M€)'] <= self.dict_data['initial investment (M€)']:
      self.dict_data['Profitability'] = 'Bad Deal'
  def ROI(self):
    """
      The Function computing the value of the Return on investment

      Parameters:
        Money Generated, initial investment
          
      Returns:
        the ROI with the formula : ((Money Generated - initial investment) / initial investment) * 100
    """ 
    self.dict_data['ROI (%)'] = ((self.dict_data['Money Generated (M€)'] - self.dict_data['initial investment (M€)']) / (self.dict_data['initial investment (M€)'])) *100

# Mandatory Method n3 #
  def data_cleaning(self):
    """
      The Function returning a dictionary of specific attributes I want to give to the director of my business
      by deleting attributes that I don't want and keeping those that I want

      Parameters:
        any data of my OOP
          
      Returns:
        Only the useful datas
    """  
    del self.dict_data['birthdate']
    del self.dict_data['total earnings (M€)']
    del self.dict_data['on-field earnings (M€)']
    del self.dict_data['off-field earnings (M€)']
    del self.dict_data['Number of Lawsuits']
    del self.dict_data['Number of Instagram followers (M)']
    del self.dict_data['Number of meetings']
    del self.dict_data['country code']
    del self.dict_data['Average Number of Instagram followers increase rate (%)']
    del self.dict_data['country']

#Let's Get access to class docstring
help(CYPartnersETL)
#Let's Get access to data_extracting method docstring
help(CYPartnersETL.data_extracting)
```

```python
# CEll Extract featuring, Buisness Life, and Director Summary #
my_clients = {} # empty dictionnary that we are going to fill in

my_unsorted_clients_list = [] # empty list that we are going to fill in with dictionnaries of clients informations
for f in lines:
  # Extract featuring part using a for loop #
  my_clients[' '.join(f.split()[:2])] = CYPartnersETL()
  my_clients[' '.join(f.split()[:2])].data_extracting(f)
  my_clients[' '.join(f.split()[:2])].data_featuring()
  # Buisness Life part using the same for loop #
  my_clients[' '.join(f.split()[:2])].average_off_field_earnings_BeforeCY()
  my_clients[' '.join(f.split()[:2])].off_field_earnings_WithCY()
  my_clients[' '.join(f.split()[:2])].off_field_earnings_increase_rate()
  my_clients[' '.join(f.split()[:2])].set_Average_numb_of_insta_followers_increase_rate(12.6) #given in the story
  my_clients[' '.join(f.split()[:2])].Numb_of_insta_follower_increase_rate_Before_CY()
  my_clients[' '.join(f.split()[:2])].Numb_of_insta_follower_increase_rate_with_CY()
  my_clients[' '.join(f.split()[:2])].set_Number_of_Meetings((random.randint(3, 15)))#given in the story
  my_clients[' '.join(f.split()[:2])].legal_expenses_before_CY()
  my_clients[' '.join(f.split()[:2])].legal_expenses_after_CY()
  my_clients[' '.join(f.split()[:2])].set_The_initial_Cost(2)
  my_clients[' '.join(f.split()[:2])].has_generated()
  my_clients[' '.join(f.split()[:2])].profitability()
  my_clients[' '.join(f.split()[:2])].ROI()
  # Director summary using the same for loop #
  my_clients[' '.join(f.split()[:2])].data_cleaning()
  my_unsorted_clients_list.append(my_clients[' '.join(f.split()[:2])].dict_data) #adding to the list

    # For the director summary, let's sort my list of client (out of the for loop) using built-in functions, module and package#
from operator import itemgetter
my_sorted_clients_list = sorted(my_unsorted_clients_list, key=itemgetter('ROI (%)'), reverse=True) 

    # For the director summary, let's filter my list of client (out of the for loop) using built-in functions#
my_almost_filtred_sorted_clients_list = list(filter(lambda clients: clients['ROI (%)'] > 0, my_sorted_clients_list))
my_totally_filtred_sorted_clients_list = list(filter(lambda clients: clients['off-field earnings increase rate (%)'] > 0, my_almost_filtred_sorted_clients_list))
```
